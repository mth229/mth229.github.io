```julia
plot(sin, 0, 2pi)		# plot(f, a, b)
```

<p>Often most of the battle is <em>judiciously</em> choosing the values of &#36;a&#36; and &#36;b&#36; so that the graph highlights a feature of interest. Such as a relative maximum or minimum, a zero, a vertical asymptote, a horizontal asymptote, a slant asymptote...</p>

<p>The use of a function as an argument is not something done with a calculator, but is very useful when using <code>julia</code> for calculus as many actions operate on the function, not the values of the function.</p>

<hr />

<p>A plot is nothing more than a connect-the-dot graph of &#36;x&#36; and &#36;y&#36; values. It can be useful to know how to do the steps. The above could be done with:</p>

```julia
a, b = 0, 2pi
xs = linspace(a, b)  		# 50 points between a and b
ys = map(sin, xs)	        # or ys = [sin(x) for x in xs] (see the notes)
plot(xs, ys)	
```

<p>The <code>xs</code> and <code>ys</code> are written as though they are "plural" because these variables contain 50 values each in a container (a vector in this case). The <code>map</code> command "maps" a function to each value in the container and returns a new container with the "mapped" values. In the example above, these are the values for the <code>sin</code> at each <code>x</code>.</p>

<p>Containers (vectors in this case) are often constructed by combining like values within square brackets separated by commas: e.g., <code>&#91;a,b&#93;</code>. For plotting, we can combine functions using <code>&#91;&#93;</code> and all will plot:</p>

```julia
plot([sin, cos], 0, 2pi)
```
